```python
import numpy as np
import time

#input variables
inputx = np.array([[1],[2],[0.5],[2]]).T
#target
targety = np.array([[0],[1],[0]]).T

def sigmoid(x):
    return  1 / (1 + np.exp(-x))

#list for saving the result
weight =[]
outputy=[]
mse=[]

start_time = time.time()

Running untuk pencarian random yang diberi bobot berdasarkan jumlah iterasi 
for iteration in range(0,10000):
    w = np.random.random((4,3))
    weight.append(w)
    bias=np.array([[0],[1],[0]]).T
    y = sigmoid(np.dot(inputx,w)+bias)
    outputy.append(y)
    minus = np.subtract(targety, outputy[iteration])
    square = np.square(minus)
    error = square.mean()
    mse.append(error)

print('MSE paling kecil disaat',iteration+1,'=', min(mse))
print("--- %s seconds ---" % (time.time() - start_time))

for i in range(0, iteration+1):
    if mse[i] == min(mse):
        index = i
print(weight[index])
```

**Metode Backward Propagation**

```python
import numpy as np
import time

#input variables
inputx = np.array([[1],[2],[0.5],[2]]).T
#target
targety = np.array([[0],[1],[0]]).T

def sigmoid(x):
    return  1 / (1 + np.exp(-x))

def fungsi_derivative(x):
    return x*(1-x)

#list for saving the result
weight =[]
outputy=[]
mse=[]

start_time = time.time()
w = np.random.random((4,3))
adjustment = np.array([[0],[1],[0]]).T

for epoch in range (0,100000):
    bias = adjustment
    y = sigmoid(np.dot(inputx,w)+bias)
    outputy.append(y)
    minus = np.subtract(targety,outputy[epoch])
    square = np.square(minus)
    error = square.mean()
    mse.append(error)
    #BackPropagation
    adjustment = minus * fungsi_derivative(y)
    w = np.dot(inputx.T, adjustment)
    weight.append(w)

print('The Smallest MSE For',epoch+1,'=', min(mse))
print("--- %s seconds ---" % (time.time() - start_time))

for i in range(0, epoch+1):
    if mse[i] == min(mse):
        index = i
print(weight[index])
```
